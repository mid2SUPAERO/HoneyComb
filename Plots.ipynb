```python

import numpy as np
import matplotlib.pyplot as plt
from honeycomb_analysis import youngs_modulus_E1,youngs_modulus_E2, poisson_ratio ,shear_modulus_G12 ,brittle_crushing_stress ,volume_surface,plastic_yield_stress,elastic_buckling_stress, E_s 
```

```python
# Constants
h = 0.1  # Height

```

```python
## E1
tl_ratios = np.linspace(0, 0.3, 31)  # Vary t/l from 0 to 0.3 with a step of 0.01
volume_percentages = []
E1_values = []
for tl_ratio in tl_ratios:
    l = h 
    t = tl_ratio * l
    if t == 0:  # Avoid division by zero
        continue
    E1 = youngs_modulus_E1(t, l, h)/E_s
    vol_percentage = volume_surface(t, l, h)
    
    E1_values.append(E1)
    volume_percentages.append(vol_percentage)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(volume_percentages, E1_values, marker='o', linestyle='-', color='b')
plt.xlabel('Volume Percentage')
plt.ylabel('Young\'s Modulus E1/Es ')
plt.title('Young\'s Modulus E1/Es as a Function of Volume Percentage')
plt.grid(True)
plt.show()






```
